```python
# @title Affichage du graphique par arrondissement
import matplotlib.pyplot as plt
import pandas as pd

# Charger le fichier CSV
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/les-arbres.csv', sep=';')

# Calculer la hauteur moyenne par arrondissement
df_grouped = df.groupby('arrondissement')['hauteurenm'].mean().reset_index()

# Tracer un graphique de la hauteur moyenne des arbres par arrondissement
plt.figure(figsize=(10, 6))
plt.bar(df_grouped['arrondissement'], df_grouped['hauteurenm'], color='skyblue')
plt.title('Hauteur moyenne des arbres par arrondissement')
plt.xlabel('Arrondissement')
plt.ylabel('Hauteur moyenne (en mètres)')
plt.xticks(rotation=90)  # Rotation des étiquettes des arrondissements pour une meilleure lisibilité
plt.grid(True)
plt.show()
```
